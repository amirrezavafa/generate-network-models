```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

def create_network(email_path, department_path):
    G = nx.Graph()
    with open(department_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 2:
                node_id, department = parts
                G.add_node(node_id, department=int(department))
    with open(email_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) == 2:
                node1, node2 = parts
                G.add_edge(node1, node2) 
    return G

def calculate_metrics(G):
    metrics = {
        'Nodes': G.number_of_nodes(),
        'Edges': G.number_of_edges(),
        'Avg Degree': np.mean([d for _, d in G.degree()]),
        'Clustering Coefficient': nx.average_clustering(G),
        'Degree Assortativity': nx.degree_assortativity_coefficient(G),
        'Avg Shortest Path': nx.average_shortest_path_length(G) if nx.is_connected(G) else -1,
    }
    return metrics

def find_best_params(G_original):
    best_params = {}
    num_nodes = G_original.number_of_nodes()

    best_m = None
    best_m_diff = float('inf')
    for m in range(1, 10):
        G_ba = nx.barabasi_albert_graph(num_nodes, m)
        diff = abs(calculate_metrics(G_ba)['Clustering Coefficient'] - calculate_metrics(G_original)['Clustering Coefficient'])
        if diff < best_m_diff:
            best_m_diff = diff
            best_m = m
    best_params["BA"] = {"m": best_m}

    best_k = None
    best_p = None
    best_ws_diff = float('inf')
    for k in range(2, 10, 2):
        for p in np.linspace(0.01, 0.5, 10):
            G_ws = nx.watts_strogatz_graph(num_nodes, k, p)
            diff = abs(calculate_metrics(G_ws)['Clustering Coefficient'] - calculate_metrics(G_original)['Clustering Coefficient'])
            if diff < best_ws_diff:
                best_ws_diff = diff
                best_k = k
                best_p = p
    best_params["WS"] = {"k": best_k, "p": best_p}

    department_counts = Counter(nx.get_node_attributes(G_original, 'department').values())
    sizes = list(department_counts.values())
    best_probs = None
    best_sbm_diff = float('inf')
    for intra_prob in np.linspace(0.05, 0.5, 5):
        for inter_prob in np.linspace(0.01, 0.1, 5):
            probs = [[intra_prob if i == j else inter_prob for j in range(len(sizes))] for i in range(len(sizes))]
            G_sbm = nx.stochastic_block_model(sizes, probs)
            diff = abs(calculate_metrics(G_sbm)['Clustering Coefficient'] - calculate_metrics(G_original)['Clustering Coefficient'])
            if diff < best_sbm_diff:
                best_sbm_diff = diff
                best_probs = probs
    best_params["SBM"] = {"sizes": sizes, "probs": best_probs}

    return best_params

def plot_degree_distribution(G, title):
    degrees = [d for _, d in G.degree()]
    plt.hist(degrees, bins=range(1, max(degrees) + 1), color='blue', alpha=0.7, rwidth=0.85)
    plt.title(f"Degree Distribution: {title}")
    plt.xlabel("Degree")
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.show()

EMAIL = 'desktop/email.txt'  
DEPARTMENT = 'desktop/email2.txt'  

G_original = create_network(EMAIL, DEPARTMENT)
best_params = find_best_params(G_original)
G_BA = nx.barabasi_albert_graph(G_original.number_of_nodes(), best_params["BA"]["m"])
G_WS = nx.watts_strogatz_graph(G_original.number_of_nodes(), best_params["WS"]["k"], best_params["WS"]["p"])
G_SBM = nx.stochastic_block_model(best_params["SBM"]["sizes"], best_params["SBM"]["probs"])
metrics_original = calculate_metrics(G_original)
metrics_BA = calculate_metrics(G_BA)
metrics_WS = calculate_metrics(G_WS)
metrics_SBM = calculate_metrics(G_SBM)

comparison = pd.DataFrame({
    "Original": metrics_original,
    "BA Model": metrics_BA,
    "WS Model": metrics_WS,
    "SBM Model": metrics_SBM
})
print("Comparison Table:")
print(comparison.T)
print("\nBest Parameters:")
print(f"BA Model - Best m: {best_params['BA']['m']}")
print(f"WS Model - Best k: {best_params['WS']['k']}, Best p: {best_params['WS']['p']}")

plot_degree_distribution(G_original, "Original Network")
plot_degree_distribution(G_BA, "BA Model")
plot_degree_distribution(G_WS, "WS Model")
plot_degree_distribution(G_SBM, "SBM")
```
