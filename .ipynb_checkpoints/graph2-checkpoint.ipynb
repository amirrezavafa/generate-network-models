```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load(file_path):
    G = nx.DiGraph()
    with open(file_path, 'r') as f:
        for line in f:
            if not line.startswith('#'):
                from_node, to_node = map(int, line.strip().split())
                G.add_edge(from_node, to_node)
    return G

WIKI = "desktop/Wiki-Vote.txt"
G_wiki = load(WIKI)

def calculate_metrics(G):
    is_directed = G.is_directed() 
    metrics = {
        "num_nodes": G.number_of_nodes(),
        "num_edges": G.number_of_edges(),
        "avg_degree": np.mean([d for n, d in G.degree()]),
        "clustering_coefficient": nx.average_clustering(G.to_undirected()),
        "degree_assortativity": nx.degree_assortativity_coefficient(G),
    }
    if is_directed:
        if nx.is_strongly_connected(G):
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G)
        else:
            largest_scc = max(nx.strongly_connected_components(G), key=len)
            G_scc = G.subgraph(largest_scc).copy()
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G_scc)
    else:
        if nx.is_connected(G):
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G)
        else:
            largest_cc = max(nx.connected_components(G), key=len)
            G_cc = G.subgraph(largest_cc).copy()
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G_cc)
    return metrics

metrics_wiki = calculate_metrics(G_wiki)
print("Metrics for Wiki-Vote Dataset:")
for key, value in metrics_wiki.items():
    print(f"{key}: {value}")

n = G_wiki.number_of_nodes()
m = 9 
k = 8  
p = 0.2
G_BA = nx.barabasi_albert_graph(n, m)
G_WS = nx.watts_strogatz_graph(n, k, p)
sizes = [2000, 2500, 2615] 
num_blocks = len(sizes)
probs = [[0.01 if i == j else 0.005 for j in range(num_blocks)] for i in range(num_blocks)]
G_SBM = nx.stochastic_block_model(sizes, probs)
metrics_BA = calculate_metrics(G_BA)
metrics_WS = calculate_metrics(G_WS)
metrics_SBM = calculate_metrics(G_SBM)

comparison = pd.DataFrame({
    "Wiki-Vote": metrics_wiki,
    "BA Model": metrics_BA,
    "WS Model": metrics_WS,
    "SBM Model": metrics_SBM
})
print("\nComparison Table:")
print(comparison.T)

def plot_degree_distribution(G, title, log_scale=True):
    degrees = [d for _, d in G.degree()]
    plt.hist(degrees, bins=50, color='blue', alpha=0.7, rwidth=0.85, log=log_scale)
    plt.title(f"Degree Distribution: {title}")
    plt.xlabel("Degree")
    plt.ylabel("Frequency (log scale)" if log_scale else "Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.show()

plot_degree_distribution(G_wiki, "Wiki Dataset")
plot_degree_distribution(G_BA, "Barabsi-Albert Model")
plot_degree_distribution(G_WS, "Watts-Strogatz Model")
plot_degree_distribution(G_SBM, "Stochastic Block Model")
```

```python
import itertools
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def load(file_path):
    G = nx.DiGraph()
    with open(file_path, 'r') as f:
        for line in f:
            if not line.startswith('#'):
                from_node, to_node = map(int, line.strip().split())
                G.add_edge(from_node, to_node)
    return G

WIKI = "desktop/Wiki-Vote.txt"
G_wiki = load(WIKI)

def calculate_metrics(G):
    is_directed = G.is_directed() 
    metrics = {
        "num_nodes": G.number_of_nodes(),
        "num_edges": G.number_of_edges(),
        "avg_degree": np.mean([d for n, d in G.degree()]),
        "clustering_coefficient": nx.average_clustering(G.to_undirected()),
        "degree_assortativity": nx.degree_assortativity_coefficient(G),
    }
    if is_directed:
        if nx.is_strongly_connected(G):
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G)
        else:
            largest_scc = max(nx.strongly_connected_components(G), key=len)
            G_scc = G.subgraph(largest_scc).copy()
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G_scc)
    else:
        if nx.is_connected(G):
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G)
        else:
            largest_cc = max(nx.connected_components(G), key=len)
            G_cc = G.subgraph(largest_cc).copy()
            metrics["avg_shortest_path"] = nx.average_shortest_path_length(G_cc)
    return metrics

def compare_metrics(metrics1, metrics2):
    similarity_score = 0
    for key in metrics1:
        if key in metrics2:
            similarity_score += abs(metrics1[key] - metrics2[key])
    return similarity_score

metrics_wiki = calculate_metrics(G_wiki)
print("Metrics for Wiki-Vote Dataset:")
for key, value in metrics_wiki.items():
    print(f"{key}: {value}")

n = G_wiki.number_of_nodes()

best_BA = {"m": None, "score": float('inf')}
for m in range(1, 10):  # Test small ranges for m
    G_BA = nx.barabasi_albert_graph(n, m)
    metrics_BA = calculate_metrics(G_BA)
    score = compare_metrics(metrics_wiki, metrics_BA)
    if score < best_BA["score"]:
        best_BA = {"m": m, "score": score}

best_WS = {"k": None, "p": None, "score": float('inf')}
for k, p in itertools.product(range(2, 10, 2), np.linspace(0.01, 0.2, 5)):
    G_WS = nx.watts_strogatz_graph(n, k, p)
    metrics_WS = calculate_metrics(G_WS)
    score = compare_metrics(metrics_wiki, metrics_WS)
    if score < best_WS["score"]:
        best_WS = {"k": k, "p": p, "score": score}

sizes = [2000, 2500, 2615]
best_SBM = {"probs": None, "score": float('inf')}
for prob in np.linspace(0.01, 0.1, 5):
    probs = [[prob if i == j else prob / 2 for j in range(len(sizes))] for i in range(len(sizes))]
    G_SBM = nx.stochastic_block_model(sizes, probs)
    metrics_SBM = calculate_metrics(G_SBM)
    score = compare_metrics(metrics_wiki, metrics_SBM)
    if score < best_SBM["score"]:
        best_SBM = {"probs": probs, "score": score}

print("\nBest Parameters:")
print(f"Barabsi-Albert: m = {best_BA['m']} with similarity score = {best_BA['score']}")
print(f"Watts-Strogatz: k = {best_WS['k']}, p = {best_WS['p']} with similarity score = {best_WS['score']}")
print(f"Stochastic Block Model: probs = {best_SBM['probs']} with similarity score = {best_SBM['score']}")
```
